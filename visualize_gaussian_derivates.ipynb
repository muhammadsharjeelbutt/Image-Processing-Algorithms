```
import math as m
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
```

# Visualize 3D Surface Plot

```

def surface_plot_3d(X,Y,Z):
  filter_size=5
  fig = plt.figure()
  ax = plt.axes(projection="3d")
  ax = fig.add_subplot(111, projection='3d')

  # transforming x,y,z into 2D arrays for plotting
  xs = X.reshape((filter_size, filter_size))
  ys = Y.reshape((filter_size, filter_size))
  zs = Z.reshape((filter_size, filter_size))

  # passing x,y,z points to plot
  ax.plot_wireframe(xs, ys, zs, color='black')
  ax.set_xlabel('x' )
  ax.set_ylabel('y' )
  ax.set_zlabel('z')
  plt.show()

  # fig = plt.figure(figsize=(20,10))
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  # 3D surface plot
  ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, cmap='binary_r', edgecolor='black')

  ax.set_title('Surface plot');
  plt.show()
```

# 1.a First Order Partial Derivative X Gaussian Kernel

```
def first_order_derivative_X_gaussian_kernel(filter_size,sigma):
# preferred sigma = 0.5
# preferred filter size = 5
  m = filter_size//2
  n = filter_size//2
  X=[]
  Y=[]
  Z=[]
  for x in range(-m, m+1):
          for y in range(-n, n+1):
              x1 = 2*np.pi*(sigma**4)
              x2 = np.exp(-(x**2 + y**2)/(2* sigma**2))
              derv_x = -(x/x1)*x2
              X.append(x)
              Y.append(y)
              Z.append(derv_x)
  
  surface_plot_3d(np.array(X),np.array(Y),np.array(Z))

first_order_derivative_X_gaussian_kernel(5,0.5)
```

```
def first_order_derivative_Y_gaussian_kernel(filter_size,sigma):
# preferred sigma = 0.5
# preferred filter size = 5
  m = filter_size//2
  n = filter_size//2
  X=[]
  Y=[]
  Z=[]
  for x in range(-m, m+1):
          for y in range(-n, n+1):
              x1 = 2*np.pi*(sigma**4)
              x2 = np.exp(-(x**2 + y**2)/(2* sigma**2))
              derv_y = -(y/x1)*x2
              X.append(x)
              Y.append(y)
              Z.append(derv_y)
  
  surface_plot_3d(np.array(X),np.array(Y),np.array(Z))

first_order_derivative_Y_gaussian_kernel(5,0.5)
```

# 2.a Second Order Partial Derivative X Gaussian Kernel

```

def second_order_derivative_X_gaussian_kernel(filter_size,sigma):
# preferred sigma = 0.5
# preferred filter size = 5
  m = filter_size//2
  n = filter_size//2
  X=[]
  Y=[]
  Z=[]
  for x in range(-m, m+1):
          for y in range(-n, n+1):
              x1 = 2*np.pi*(sigma**6)
              x2 = np.exp(-(x**2 + y**2)/(2* sigma**2))
              num = (x**2-sigma**2)
              derv_x = (num/x1)*x2
              X.append(x)
              Y.append(y)
              Z.append(derv_x)
  surface_plot_3d(np.array(X),np.array(Y),np.array(Z))

second_order_derivative_X_gaussian_kernel(5,0.5)
```
